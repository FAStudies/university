# Polynomial Curve Fitting and Basics of Neurons

**Linear Fit:**

$$ W_1 + W_0 = y $$

**Polynomial Curve Fit:**
$$ W_3x^3 + W_2x^2 + W_1x + W_0 = y $$

**Linear Regression:**
\begin{equation} \begin{bmatrix} x_1 & 1\\\ x_2 & 1 \\\ x_3 & 1 \\\ \cdots & \cdots \\\ x_n & 1 \end{bmatrix} * \begin{bmatrix} W_1 \\\ W_0 \end{bmatrix} = \begin{bmatrix} y_1 \\\ y_2 \\\ y_3 \\\ \cdots \\\ y_n \end{bmatrix} \end{equation}

\begin{equation} X * W = Y  \end{equation}

\begin{equation} X^TX * W = X^TY  \end{equation}

\begin{equation} W = (X^TX)^{-1}X^T Y  \end{equation}

\begin{equation} \begin{bmatrix} x_1^2 & x_1 & 1\\\ x_2^2 & x_2 & 1 \\\ x_3^2 & x_3 & 1 \\\ \cdots & \cdots & \cdots \\\ x_n^2 & x_n & 1 \end{bmatrix} * \begin{bmatrix} W_2 \\\ W_1 \\\ W_0 \end{bmatrix} = \begin{bmatrix} y_1 \\\ y_2 \\\ y_3 \\\ \cdots \\\ y_n \end{bmatrix} \end{equation}

$$ (X^TX) $$

\begin{equation} x = \begin{bmatrix} 1 & x_1 \\\ 1 & x_2 \\\ 1 & x_3 \\\ \cdots & \cdots \\\ 1 & x_n \end{bmatrix} * \begin{bmatrix} W_0 \\\ W_1 \end{bmatrix} = \begin{bmatrix} y_1 \\\ y_2 \\\ y_3 \\\ \cdots \\\ y_n \end{bmatrix} \end{equation}

\begin{equation} \begin{bmatrix} 1 & 1 & 1 & \cdots & 1 \\\ x_1 & x_2 & x_3 & \cdots & x_n \end{bmatrix} * \begin{bmatrix} 1 & x_1 \\\ 1 & x_2 \\\ 1 & x_3 \\\ \cdots & \cdots \\\ 1 & x_n \end{bmatrix} \end{equation}

$$ (2 \times n)   \times   (n \times 2) $$

\begin{equation} \begin{bmatrix} \sum_{i=1}^n 1 & \sum_{i=1}^n x_i \\\ \sum_{i=1}^n x_i & \sum_{i=1}^n x_i^2 \end{bmatrix} * \begin{bmatrix} W_0 \\\ W_1 \end{bmatrix} = X^TY \end{equation}


$$ (X^TY) $$

\begin{equation} = \begin{bmatrix} 1 & 1 & 1 & \cdots & 1 \\\ x_1 & x_2 & x_3 & \cdots & x_n \end{bmatrix} * \begin{bmatrix} y_1 \\\ y_2 \\\ y_3 \\\ \cdots \\\ y_n \end{bmatrix} \end{equation}

\begin{equation} = \begin{bmatrix} \sum_{i=1}^n y_i \\\ \sum_{i=1}^n x_iy_i \end{bmatrix}\end{equation}

\begin{equation} W = \begin{bmatrix} W_0 \\\ W_1 \end{bmatrix} = \begin{bmatrix} N & \sum x_i \\\ \sum x_i & \sum n_i^2 \end{bmatrix} \begin{bmatrix} \sum_{i=1}^n y_i \\\ \sum_{i=1}^n x_iy_i \end{bmatrix} \end{equation}

\begin{equation} \begin{bmatrix} W_2 \\\ W_1 \\\ W_0 \end{bmatrix} = \begin{bmatrix} N & \sum x_i & \sum x_i^2 \\\ \sum x_i & \sum x_i^2 & \sum x_i^3 \\\ \sum x_i^2 & \sum x_i^3 & \sum x_i^4 \end{bmatrix} \begin{bmatrix}\sum y_i \\\ \sum x_iy_i \\\ \sum x_i^2y_i \end{bmatrix} \end{equation} 

Redo:

\begin{equation} W = (X^TX)^{-1}X^T Y  \end{equation}

\begin{equation} \begin{bmatrix} W_0 \\\ W_1 \end{bmatrix} = \begin{bmatrix} \begin{bmatrix} 1 & 1 & 1 & \cdots & 1 \\\ x_1 & x_2 & x_3 & \cdots & x_n \end{bmatrix} * \begin{bmatrix} 1 & x_1 \\\ 1 & x_2 \\\ 1 & x_3 \\\ \cdots & \cdots \\\ 1 & x_n \end{bmatrix} \end{bmatrix} ^ {-1} (X^TY) \end{equation}

\begin{equation} = \begin{bmatrix} \sum_{i=1}^n 1 & \sum_{i=1}^n x_i \\\ \sum_{i=1}^n x_i & \sum_{i=1}^n x_i^2 \end{bmatrix}  (X^TY)
\end{equation}

\begin{equation} = \begin{bmatrix} \sum_{i=1}^n 1 & \sum_{i=1}^n x_i \\\ \sum_{i=1}^n x_i & \sum_{i=1}^n x_i^2 \end{bmatrix}  \begin{bmatrix} \begin{bmatrix} 1 & 1 & 1 & \cdots & 1 \\\ x_1 & x_2 & x_3 & \cdots & x_n \end{bmatrix} * \begin{bmatrix} y_1 \\\ y_2 \\\ y_3 \\\ \cdots \\\ y_n \end{bmatrix}\end{bmatrix}
\end{equation}

\begin{equation} \begin{bmatrix} W_0 \\\ W_1 \end{bmatrix} = \begin{bmatrix} N & \sum x_i \\\ \sum x_i & \sum n_i^2 \end{bmatrix} \begin{bmatrix} \sum_{i=1}^n y_i \\\ \sum_{i=1}^n x_iy_i \end{bmatrix} \end{equation}

Similarly, for Quadritic Polynomial:
\begin{equation} \begin{bmatrix} W_2 \\\ W_1 \\\ W_0 \end{bmatrix} = \begin{bmatrix} N & \sum x_i & \sum x_i^2 \\\ \sum x_i & \sum x_i^2 & \sum x_i^3 \\\ \sum x_i^2 & \sum x_i^3 & \sum x_i^4 \end{bmatrix} \begin{bmatrix}\sum y_i \\\ \sum x_iy_i \\\ \sum x_i^2y_i \end{bmatrix} \end{equation}
