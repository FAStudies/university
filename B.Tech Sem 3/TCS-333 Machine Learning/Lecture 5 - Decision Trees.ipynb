## Entropy
Randomness in data. <br>


$ x => \mbox{probability(E)}  = P(E)$
$ E(x) = -\sum_{i=1}^n P(x_i)log_2P(x_1) $$    
(Shanon's Entropy)<br>

**Example:** Tossing a fair coin:
$ \mbox{Probability of Head} = P(H) = \frac{1}{2} = P(T) $    
$ \mbox{Entropy} = E(x) = -\big( \frac{1}{2} log_2 \big( \frac{1}{2} \big) + log_2 \big( \frac{1}{2} \big) \big) $    
$ = \big( \frac{1}{2} (-1) + \frac{1}{2} (-1) \big) $    
$ = -(-1) = 1 $     

**More entropy = More randomness = Difficult to predict output**

**Example:** Tossing an unfair coin:     
$ P(H) = 0.7,  P(T) = 0.3 $    
$ \mbox{Entropy} = E(x) = - ( 0.7 log_2 0.7) + 0.3 log_2 0.3) $    
$ = 0.881 (< 1) $    
It is easier to predict output here as there are more chances of getting head than there are of getting tail.
