# Hierarchial Clustering

## Types of Clustering:
1. **Agglomerative:** Bottom-up Approach
2. **Divisive:** Top-down Approach

## Divisive Clustering
**Top-down approach**    
Starts with bigger clusters & divides them into smaller clusters.!

## Agglomerative Clustering
**Bottom-up approach**    
Starts with small clusters & merge them together to create bigger clusters.

### Advantages of Agglomerative Clustering
1. We need not presume the number of clusters. 
2. It may give better meaningful classification. 

### Disadvantages of Agglomerative Clustering
1. We cannot divide the clusters once they are merged. 
2. Slow for large Dataset

### Algorithm
1. Data should be in numerical form. 
2. Find distance of each point from all other points. 
3. Merge the closest two clusters into single clusters. 
4. Put the value in dendogram.
5. Repeat 2-3 until only 1 cluster is left. 

## Dendogram
We can use dendograms to visualise the grouping history and figure out the optimal number of clusters.

1. Determine the largest vertical distance that does not intersect with any other cluster.
2. Draw horizontal lines at both eextremeties.
3. The optimal number of clusters is equal to the number of vertical lines going through.

## Measuring Distance
There are two ways of measuring the distance:
1. **Euclidean Distance:** $ \sqrt{ (x_1 - x_2)^2 + (y_1 - y_2)^2 } $
2. **Manhattan Distance:** $ |x_1-x_2| + |y_1-y_2| $


## Linkage
Linkage is the distance between two clusters.

### Types of Linkage:
1. Single Linkage

### Single Linkage:
Distance between two clusters is the distance between their two closest point.

$$ D(c1,c2) = \mbox{MIN}(D(x_{c_{i}},x_{c_{j}})) $$    
$$ x_{c_{i}} \in c_1,x_{c_{j}} \in c_2 $$    

### Complete Linkage
Distance between two clusteers is the max distance between any of their points.


$$ D(c1,c2) = \mbox{MAX}(D(x_{c_{i}},x_{c_{j}})) $$    
$$ x_{c_{i}} \in c_1,  x_{c_{j}} \in c_2 $$ 

### Average Linkage
Distance between two clusteers is the average of all the distance between all their points.


$$ D(c1,c2) = \mbox{AVG}(D(x_{c_{i}},x_{c_{j}})) $$ 
$$ = D(c1,c2) = \frac{1}{n_{c_1} * n_{c_1}} \sum_{i=1}^{n_{c_1}} \sum_{j=1}^{n_{c_2}} (D(x_{c_{i}},x_{c_{j}})) $$
$$ x_{c_{i}} \in c_1, x_{c_{j}} \in c_2 $$
