```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as hc
from scipy.cluster.hierarchy import dendrogram, ward
```

```python
cust_data = pd.read_csv('customers.csv')
print(cust_data.head(12))
data = cust_data.iloc[:, 3:5].values
```

```python
import math

def getDistance(a, b):
    return round(math.sqrt(
            ((float(a[0])-float(b[0]))**2
            +
            ((float(a[0])-float(b[0]))**2
            )
            )
        ), 2)

def getMinimium(distance_matrix):
    p = [0,0]
    mn = distance_matrix[0][0]
    for i in range(0, len(distance_matrix)):
        for j in range(0, len(distance_matrix[i])):
            if (distance_matrix[i][j]>0 and distance_matrix[i][j]<mn):
                mn = distance_matrix[i][j]
                p=[i,j]     
    return p

def newPoint(pt):
    # Creates a new point using average linkage
    x = (float(pt[0][0]) + float(pt[1][0]))/2
    y = (float(pt[0][1]) + float(pt[1][1]))/2
    return x,y

def fit(data):
    
    l=0
    names={}
    
    while(len(data)>1):
        l+=1
        matrix=np.array([])
        print('Distance Matrix No. ', l, ':')
        for i in range(0, len(data)):
            m=[]
            for j in range(0, len(data)):
                m.append(0)
            for j in range(0, len(data)):
                m[j] = getDistance(data[i], data[j])
            print(str(m))
            np.append(matrix,m)
            
        pts=np.array([])

        minimum = getMinimium(data)

        p1 = data[minimum[0]]
        np.append(pts,p1)
        data = np.delete(data, minimum[0])


        p2 = data[minimum[1]-1]
        np.append(pts,p2)
        data = np.delete(data, minimum[1]-1)

        midpoint = newPoint(pts)
        np.append(data, midpoint)
        c1 = names.pop(str(p1))
        c2 = names.pop(str(p2))
        
        names[str(midpoint)] = "["+str(c1)+str(c2)+"]"
        outline=names[str(midpoint)]
        
    print('Cluster is: ', names[str(midpoint)])
```

```python
fit(data)
```
